```
def merge_seats(x):
    if 2 <= x <= 4:
        return '2-4'
    elif x > 5:
        return '>5'
    else:
        return '5'
```

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime

def preprocess_and_feature_engineer(df_input):
    df = df_input.copy()

    df = df.loc[df.mileage != 0].copy()

    outlier_threshold = df.selling_price.quantile(0.95)
    df.loc[df.selling_price > outlier_threshold, 'selling_price'] = outlier_threshold

    current_year = datetime.now().year
    df['age'] = current_year - df['year']
    df = df.drop(columns=['year'])

    if 'full_name' in df.columns:
        df['make'] = df.full_name.apply(lambda x: x.split()[0])
        df['model'] = df.full_name.apply(lambda x: " ".join(x.split()[1:]))
        df = df.drop("full_name", axis=1)

    df['seats'] = df['seats'].astype(float)
    df['seats'] = df['seats'].apply(merge_seats)

    categorical_cols_to_encode = ['seller_type', 'fuel_type', 'transmission_type', 'seats']

    for col in categorical_cols_to_encode:
        if col in df.columns and df[col].dtype == 'object':
            dummies = pd.get_dummies(df[col], prefix=col)
            if dummies.shape[1] > 1:
                df = pd.concat([df, dummies.iloc[:, 1:]], axis=1)
            else:
                df = pd.concat([df, dummies], axis=1)
            df = df.drop(columns=[col])

    df['make'] = df.groupby('make')['selling_price'].transform('median')
    df['model'] = df.groupby('model')['selling_price'].transform('median')

    numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
    scaler = MinMaxScaler()
    df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

    return df
```
